```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
from sympy import symbols, Eq, solve, diff

xs = np.array([0,1,2,3,4])
ys = np.array([0,1,1,2,2])

a,b = symbols('a, b')

def f(x):
   return a*x + b

phi = 0
for i in range (xs.size):
    phi += (f(xs[i]) - ys[i]) ** 2

phi
    
```

```python
diff(phi, a)
```

```python
sol = solve([Eq(diff(phi,a), 0),
             Eq(diff(phi,b), 0)])

sol
```

```python
def f_sol(x,sol):
    return sol[a] * x + sol[b]

x_axis = np.linspace(0, 4, 50)
plt.plot(x_axis, f_sol(x_axis, sol), 'red')
plt.scatter(xs, ys)
plt.show()
```

```python
import pandas as pd
df = pd.read_csv('CO_2_data.csv')
years = df['year']
co2 = df['co2']
```

```python
a,b = symbols('a, b')

def f(x):
    return a*x + b

phi = 0
for i in range(years.size):
    phi += (f(years[i]) - co2[i]) ** 2

sol = solve([Eq(diff(phi, a), 0),
            Eq(diff(phi, b), 0)])

def f_sol(x, sol):
    return sol[a] * x + sol[b]

x_axis = np.linspace(1940, 2022, 1000)
plt.plot(x_axis, f_sol(x_axis, sol), 'red')
plt.scatter(years, co2)
plt.show()
```

```python
df = pd.read_csv('wind_turbine_data_sample.csv')
wind = df['wind speed']
power = df['power output']
wind
```

```python
a,b,c,d = symbols('a, b, c, d')

def f(x):
    return a * x ** 3 + b * x ** 2 + c * x + d

phi = 0 
for i in range (wind.size):
    phi += (f(wind[i]) - power[i]) ** 2

sol = solve([Eq(diff(phi,a), 0),
             Eq(diff(phi,b), 0),
             Eq(diff(phi,c), 0),
             Eq(diff(phi,d), 0)])

def f_sol(x, sol):
    return sol[a]*x**3 + sol[b]*x**2 + sol[c]*x + sol[d]

x_axis = np.linspace(5, 13, 100)
plt.plot(x_axis, f_sol(x_axis, sol), 'red')
plt.scatter(wind, power)
plt.show()
```

```python
df = pd.read_csv('amazon_sales_net_revenue.csv')
years = df['year']
revenue = df['revenue']
```

```python
b,c = symbols('b, c')

def f(x):
    return b * x + c

phi = 0
for i in range(years.size):
    phi += (f(years[i]) - np.log(revenue[i])) ** 2

sol = solve([Eq(diff(phi, b), 0),
             Eq(diff(phi, c), 0)])

def exp_sol(x,sol):
    return np.exp(float(sol[b])*x + float(sol[c]))

x_axis = np.linspace(2004, 2021, 100)
plt.plot(x_axis, exp_sol(x_axis, sol), 'red')
plt.scatter(years, revenue)
plt.show()

print(sol)
```
